## Модуль 3. Циклы

В python существует только два вида циклов: ```for...in``` и ```while```. Цикл ```for``` удобен для работы с перечислениями, ```while``` - для повтора кода, пока истинно то или иное условие.

### Цикл for

Вернее, ```for ... in```. Код этого вида циклов можно прочитать как "для каждого элемента в последовательности выполнить некоторый код". После слова ```in``` можно указать любой объект, который можно итерировать, то есть "перебирать". Как именно эти объекты работают, и какие объекты можно передавать в цикл ```for```, вы узнаете позже. Рассмотрим, как ```for``` работает с функцией ```range()```.

Функция ```range(begin, end, step)``` генерирует последовательность целых чисел в диапазоне от ```begin``` до ```end``` (не включая ```end```), с шагом ```step```. Таким образом, мы можем использовать эту функцию для того, чтобы некоторый код выполнился заданное количество раз.

В самом простом виде, код для цикла ```for ... in range()```, выглядит так:
```python
<код программы>

for i in range(n):
    <блок кода, который должен выполниться n раз>
    

<продолжение кода программы>
```

При этом счетчик ```i``` (его, кстати, можно именовать как угодно, не обязательно ```i```) будет принимать значения ```0, 1, 2, ..., n-1```. \
__ВНИМАНИЕ__: значения счетчика всегда включают нижнюю границу (в данном случае, ```0```), и никогда - верхнюю.

Если нам нужно, чтобы счетчик ```i``` принимал значения в заданном диапазоне - функцию ```range()``` можно вызвать с двумя параметрами: ```begin``` и ```end```:   
```python
for i in range(begin, end):
    # i будет принимать значения от begin до end-1
    <блок кода, который должен выполняться в цикле begin-end раз>
```

Еще можно задать шаг, с которым должна генерироваться последовательность ```i``` для цикла, для этого ```range()``` нужно вызвать с тремя параметрами: ```begin```, ```end``` и ```step```.  

````python
print("Счетчик в цикле for, когда range() возвращает 5 чисел от 0 до 4:")
for i in range(5): 
    print (i)

print("Счетчик в цикле for, когда range(1,6) возвращает 5 чисел от 1 до 5:")
for i in range(1,6): 
    print (i)

print("Счетчик в цикле for, когда range(1,7,2) возвращает числа от 1 до 5 с шагом 2:")
for i in range(1,7,2):
    print (i)
    
````

### Цикл while

Выполнять код внутри блока, пока условие, которое записано после оператора ```while```, истинно.

```python
<код программы>

while <какое-либо условие>:
    <код, который должен выполниться столько раз, сколько позволяет указанное условие>
    

<продолжение кода программы>
```

Иногда в ```while``` добавляют условие ```else:```, код после которого должен выполниться только если условие выполнения цикла ```while``` изначально ложно.

````python
a = 4
b = 0
while b < a :
    print(b)
    b += 1
else:
    print("b is greater than a")
````

### Прерывание циклов

Два вида прерываний:
1. ```break``` - выход из цикла. Полезен, например, при поиске, когда искомая позиция найдена - зачем листать дальше?
2. ```continue``` - возврат к началу. Применимо, если надо пропустить какие-то позиции при переборе значений.

````python
print("Счетчик в цикле for, когда range() возвращает 5 чисел от 0 до 4 (выход на позиции 3):")
for i in range(5): 
    if i==3:
        break;
    print (i)
````

````python
print("Счетчик в цикле for, когда range() возвращает 5 чисел от 0 до 4 (пропуск позиции 3):")
for i in range(5): 
    if i==3:
        print(f"No {i}!")
        continue;
    print (i)
````
