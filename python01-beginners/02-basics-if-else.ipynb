### Переменные в Python

Переменная в Python создается и инициализируется в момент присваивания ей значения. Вывести значение переменной на экран можно функцией ```print()```:

````python
some_var = "Hello, world!"
print(some_var)
````

Переменной также можно присваивать не только конкретное значение, но и значение другой переменной:

````python
another_var = some_var
print(another_var)
````

Над переменными можно выполнять различные операции (например, арифметические), их можно передавать в функции, а также присваивать им значения, которые возвращают функции:

````python
A = 10
B = 5
print( (A + B) / 2 )  # среднее арифметическое A и B

C = A + B # переменной С присвоен результат операции сложения
print(C)

D = (A + B) / 2 # D - среднее арифметическое A и B
print(D)

E = round(D) # мы передали переменную D в функцию round(), и затем значение, возвращаемое ей присовили переменной E
print(E)
````

### Операции сравнения, логические операторы

Сравнение: 
- ```==``` - равенство
- ```!=``` - неравенство
- ```>```, ```<```  - больше, меньше
- ```>=```, ```<=```  - больше или равно, меньше или равно

Эти операции могут использоваться между переменными разных типов и могут быть определены для любого вашего класса.

Логические операторы: ```and``` - "и", ```or``` - "или", ```not``` - "не", логическое отрицание.  Приоритет выполнения этих операторов - сначала "отрицание", потом "и", потом "или".

````python
# можно сравнивать переменные любого типа
var_int = 5
var_float = 5.0
var_str = "5"

print( var_int==var_float ) # True
print( var_str==var_int ) # False
print( var_str==str(var_int) ) # Но если привести тип "вручную", то True

var_str_50 = "50"
print( var_int > var_float ) # False
print( var_str_50 > var_str ) # True, но они сравниваются посимвольно:
var_str_6 = "6"
print( var_str_50 > var_str_6 ) # False, ASCII-код символа "6" > "5"

# логические операторы
print( True or False ) # "или" - "or"
print( True and False ) # "и" - "and"

````

Также есть побитовые логические операторы: ```&``` - "и", ```|``` - "или", ```^``` - "исключающее или" (XOR) и ```~``` - "не", отрицание. Они выполняются для каждого бита данных в ваших переменных.

````python
a = 0b01001010
b = 0b11010101

print(bin(a & b))
print(bin(a | b))
print(bin(a ^ b))
print(bin(~a))
````

Все как в жизни и любых других языках программирования. Приоритет операций в сложных выражениях такой же, как и в реальной жизни: сначала возведение в степень, потом умножение/деление, потом сложение/вычитание. Для ряда операций можно использовать модуль math.

#### Инкременты/декременты: арифметические, геометрические, экспоненциальные

Чтобы не переписывать имена переменных в коде, который модифицирует их самих, в Python есть операция инкремента:

````python
counter = 1
print(counter)

counter = counter + 1 # увеличить переменную counter на 1
print(counter)

counter += 1 # можно написать короче через операцию инкремента
print(counter)

# инкремент работает для любых арифметических операций
counter **= 2
print(counter)
````

### Ветвления, if...elif...else, блоки кода

Конструкция условного оператора в Python почти ничем не отличается от аналогичных в других языках программирования: после оператора ```if``` следует выражение, которое должно возвращать булево значение (```True``` или ```False```). Если оно возвращает истинное значение, происходит выполнение кода внутри __блока__, который следует за данным условием:
```python
if <какое-либо условие>:
    # начало блока
    <1-я строка кода, который должен выполниться>
    <2-я строка кода, который должен выполниться>
    ...
    <n-я строка кода, который должен выполниться> # конец блока
    
    
<строка кода, который выполнится в любом случае>
<еще строка кода, который выполнится в любом случае>
```

Блоки кода в Python не имеют явного начала или окончания, как в других языках программирования (например, в C-подобных языках блоки кода выделяются скобками ```{}```, в Pascal это ```begin...end```). Блоки в Python выделяются отступами в 4 пробела. 

Новый блок начинается с отступа в 4 пробела вправо, заканчивается - отступом в 4 пробела влево (то есть мы как бы возвращаемся на предыдущий уровень отступа).

Допустимо использование табуляции, но смешивать в одном файле два этих способа нельзя! Jupyter Notebook автоматически проставляет 4 пробела, даже если вы используете клавишу Tab для выравнивания отступов.

Произвольно расставлять отступы в Python нельзя. Все строки внутри одного блока должны иметь одинаковый отступ.


````python
if 2 * 2 == 4:
    print("дважды два четыре")
    print("дважды два четыре")
    print("Это всем известно в целом мире")
    
````

Если какой-то код должен выполняться в противном случае, мы можем дополнить условный оператор оператором ```else```:

````python
A = 2
B = 2
if A + B == 4:
    print("четыре!")
else:
    print("не четыре!")
````

Если же код должен выполняться в том случае, когда выражение в начальном условии ```if``` ложно, но выполняется другое условие, то следует использовать оператор ```elif```, в котором как раз и следует прописать данное условие:

````python
season = input("Введите время года: ")
if season=='лето':
    print( 'Вот оно какое, наше лето!' )
elif season=='осень':
    print( 'Унылая пора, очей очарованье!' )
elif season=='зима':
    print( 'Зима, крестьянин торжествуя...' )
elif season=='весна':
    print( 'Весна, весна на улице, весенние деньки!' )
else:
    print( f'Введенное время года "{season}" нам неизвестно' )
        
````

Условные операторы можно вкладывать друг в друга, но уровни вложенности (уровни отступа блоков) должны соответствовать  вашим операторам:

````python
A = 2
B = 2
if A * B == 4:
    print("четыре!")
    if A==2:
        print("это же дважды два!")
else:
    print("это не четыре")
    if A==2:
        print("это не дважды два...")
````

#### If... одной строкой

Иногда бывает необходимость вставить условие на этапе присвоения значения, не выводя код в отдельны блок, как это реализовано в языках C/C++, Java, C#, JavaScript, PHP: ```x = (conditionTrue ? value_when_True : value_when_False)```

В Python эта конструкция выглядит так: ```x = value_when_True if conditionTrue else value_when_False```

````python
# if... одной строкой, аналог 
z_order = 15
xa = 1 if z_order==15 else 0
print(xa)
````
